```python
from tkinter import *
from tkinter import filedialog
import pygame
import pyaudio
import wave
import os.path

font = ('Verdana', 11)


class MusicPlayer:

    def __init__(self, window):

        self.music_file = False
        self.recorded_audio = False
        self.do_record = False

        window.geometry('530x150')
        window.title('Speech Comparator')
        window.resizable(0, 0)
        load = Button(window, text='Load Sound', width=10, font=font, command=self.load)
        play = Button(window, text='Native Speaker (Play)', width=20, font=font, command=self.play)
        stop = Button(window, text='Stop', width=10, font=font, command=self.stop)
        record_audio = Button(window, text='Record Audio', font=font, command=self.record)
        play_recorded_audio = Button(window, text='Play Recorded Audio', font=font, command=self.play_recorded)

        load.place(x=20, y=20)
        play.place(x=150, y=20)
        stop.place(x=390, y=20)
        record_audio.place(x=380, y=70)
        play_recorded_audio.place(x=160, y=70)

        self.recorded_audio = "sound.wav"

    def record(self):
        pygame.mixer.music.unload()

        FORMAT = pyaudio.paInt16
        CHANNELS = 2
        RATE = 44100
        CHUNK = 1024
        RECORD_SECONDS = 3

        audio = pyaudio.PyAudio()

        print("recording...")
        # start Recording
        stream = audio.open(format=FORMAT, channels=CHANNELS,
                            rate=RATE, input=True,
                            frames_per_buffer=CHUNK)
        frames = []
        for i in range(0, int(RATE / CHUNK * RECORD_SECONDS)):
            data = stream.read(CHUNK)
            frames.append(data)

        # stop Recording
        stream.stop_stream()
        stream.close()
        audio.terminate()
        print("recording completed")

        wave_file = wave.open(self.recorded_audio, 'wb')
        wave_file.setnchannels(CHANNELS)
        wave_file.setsampwidth(audio.get_sample_size(FORMAT))
        wave_file.setframerate(RATE)
        wave_file.writeframes(b''.join(frames))
        wave_file.close()

    def load(self):
        self.music_file = filedialog.askopenfilename()

    def play_recorded(self):
        if os.path.isfile(self.recorded_audio):
            pygame.mixer.init()
            pygame.mixer.music.load(self.recorded_audio)
            pygame.mixer.music.play()

    def play(self):
        pygame.mixer.init()
        pygame.mixer.music.load(self.music_file)
        pygame.mixer.music.play()

    def stop(self):
        pygame.mixer.music.stop()
        pygame.mixer.music.unload()


root = Tk()
app = MusicPlayer(root)
root.mainloop()
```
